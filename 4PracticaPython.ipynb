#### MATRIZ NXN CON VALORES ENTERPS ALEATORIOS ENTRE 0 Y 10

```python
n= int(input('Introduzca el tamaño de la matriz:'))
matriz= np.random.randint(0,10, size=(n,n))
print(matriz)
```

#### CALCULAR EL MINIMO Y MÁXIMO 

```python
print('El número máximo es:', np.max(matriz))
print('El número minimo es:', np.min(matriz))
```

#### CALCULAR NUEVA MATRIZ CON EL MINIMO Y MÁXIMO DE CADA FILA

```python


maximo= np.max(matriz, axis=1)
minimo= np.min(matriz, axis=1)

matriz_resultado = np.empty((n, 2))
matriz_resultado[:, 0] = maximo
matriz_resultado[:, 1] = minimo

print(matriz_resultado)
```

#### CUANTAS VECES APARECE CADA NUMERO

```python
np.unique(matriz_resultado)
conteo= np.unique(matriz_resultado, return_counts=True)
print(conteo)
```

```python
valores, conteos = np.unique(matriz_resultado, return_counts=True)
conteos_lista = conteos.tolist()
for valor, conteo in zip(valores, conteos_lista):
    print("El número", valor, "aparece", conteo, "veces en la matriz.")
```
