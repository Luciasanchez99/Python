3. Reestructurar el DataFrame tal que aparezca en una columna nueva el día de la medida (correspondiente a la columna "D..") y en otra columna el dato del contaminante.

La lista final de campos será: ["ESTACION","MAGNITUD","ANO","MES","DIA","VALOR"]


```python
import pandas as pd

datosprueba=datostotales3

columnas_dias = datosprueba.columns[datosprueba.columns.str.startswith('D')]

reestructurado = datosprueba.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], 
                                     value_vars=columnas_dias, 
                                     var_name='DIA', 
                                     value_name='VALOR')
print(reestructurado)
```

```python
reestructurado['DIA'].unique()
```

```python
reestructurado['DIA'] = reestructurado['DIA'].str.replace('D', '')
reestructurado
```

4. Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
fechas = []
for _, row in reestructurado.iterrows():
    try:
        fecha = datetime(int(row['ANO']), int(row['MES']), int(row['DIA']))
        fechas.append(fecha)
    except ValueError:
        fechas.append(None)


reestructurado['FECHA'] = fechas

reestructurado
```

5. Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones contaminantes y fecha.

```python
reestructurado = reestructurado.drop(reestructurado[np.isnat(reestructurado.FECHA)].index)
reestructurado
```
